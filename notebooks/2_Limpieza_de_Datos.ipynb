## Corrigiendo Error de Columnas Duplicadas

La tabla personajes tiene dos columnas iniciales: Tamaño del vehiculo y personje.
El resto de tabla no tiene esa division de tamaño. 

Las cuatro tablas estaban contenidas dentro de la primera. Para no romper el formato, las ultimas tres tablas tenian sus primeras dos columnas combinadas.

**Objetivo:**
- Convertir las dos primeras columnas de cada tablas en una sola.

```python
df_vehiculos = df_vehiculos.drop('Body.2', axis=1)
df_ruedas = df_ruedas.drop('Tires.2', axis=1)
df_planeadores = df_planeadores.drop('Glider.2', axis=1)
```
